```python
data=pd.read_csv("Iris.csv", header=None, names=['sepal length', 'sepal width', 'petal length', 'petal width', 'Species'])
print(data.head(5))
X=data.drop('Species',axis=1)
print(X.head(5))
Y=data['Species']
print(Y.head(5))
```

##SPLIT THE DATA SET-

```python
# Ensure the data, X, and Y variables from the corrected data loading are used here
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=30, random_state=42)
```

#Scale The Data-
 scale feature is to ensure that all are in similar scale

```python
Scalar = StandardScaler()
X_train = Scalar.fit_transform(X_train)
X_test = Scalar.transform(X_test)
```

#Train the knn model:-

```python
k=5
KNN=KNeighborsClassifier(n_neighbors=k)
KNN.fit(X_train,Y_train)
```

#Make prediction and evaluate the model-

```python
# Train the KNN model with the optimal k value (assuming the previous cell found the optimal k)
# Replace 'optimal_k' with the actual optimal k value found from the previous cell's output
# For now, I will use k=5 as defined earlier, but you should update this with the optimal k.
# If you want to use the optimal k found in the previous cell, you would need to store it in a variable.
# Let's assume for this fix that the optimal k is determined and used before this cell.
# However, since the previous cell only plots the accuracies, let's refit with k=5 as it was before.
# To use the optimal k from the plot, you would need to find the k corresponding to the max accuracy.

# Refitting the KNN model with k=5 (or replace with the optimal k from the plot)
k = 5 # Or the optimal k found from the previous cell
KNN = KNeighborsClassifier(n_neighbors=k)
KNN.fit(X_train, Y_train)


Y_pred=KNN.predict(X_test)
print("Accuracy:", accuracy_score(Y_test,Y_pred))

print("confusion_Matrix:",
confusion_matrix(Y_test,Y_pred))
print("classification Report: \n",classification_report(Y_test,Y_pred))
```

#optimized the k value:-

```python
from sklearn.model_selection import cross_val_score

accuracies=[]

for k in range(1,21):
    KNN=KNeighborsClassifier(n_neighbors=k)
    scores=cross_val_score(KNN,X_train,Y_train,cv=5)
    accuracies.append(scores.mean())

plt.plot(range(1,21), accuracies)
plt.xlabel("Cross Validation Accuracy")
plt.ylabel("optimum k value selection")
plt.show()
```
